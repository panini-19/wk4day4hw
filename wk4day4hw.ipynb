```python
# # #FIRST POKEMON: PIKACHU:


import requests
url = 'https://pokeapi.co/api/v2/pokemon/pikachu'
response = requests.get(url)
response.status_code

poke_data = response.json()
pikachu_data = poke_data

def get_pika_info(data):
    pika_data = []
    for pika in data:
        pika_dict = {
            'pokemon_name': data['forms'][0]['name'],
            'ability_name': data['abilities'][1]['ability']['name'],
            'base_experience': data['base_experience'],
            'attack_base_stat': data['stats'][1]['base_stat'],
            'hp_base_stat': data['stats'][2]['base_stat'],
            'defense_base_stat': data['stats'][2]['base_stat'],
            'sprite': data['sprites']['front_shiny']
        }

        pika_data.append(pika_dict)
        return pika_data
get_pika_info(pikachu_data)
print(get_pika_info(pikachu_data))

# # # NEXT POKEMON: MEWTWO


import requests
url = 'https://pokeapi.co/api/v2/pokemon/mewtwo'
response = requests.get(url)
response.status_code

poke_data = response.json()
mewtwo_data = poke_data

def get_mew_info(data):
    mew_data = []
    for mew in data:
        mew_dict = {
            'pokemon_name': data['forms'][0]['name'],
            'ability_name': data['abilities'][1]['ability']['name'],
            'base_experience': data['base_experience'],
            'attack_base_stat': data['stats'][2]['base_stat'],
            'hp_base_stat': data['stats'][2]['base_stat'],
            'defense_base_stat': data['stats'][2]['base_stat'],
            'sprite': data['sprites']['front_shiny']
        }

        mew_data.append(mew_dict)
        return mew_data
get_mew_info(mewtwo_data)
print(get_mew_info(mewtwo_data))

# # #NEXT POKEMON: LUCARIO
import requests
url = 'https://pokeapi.co/api/v2/pokemon/lucario'
response = requests.get(url)
response.status_code

poke_data = response.json()
lucario_data = poke_data

def get_luc_info(data):
    lucario_data = []
    for luc in data:
        luc_dict = {
            'pokemon_name': data['forms'][0]['name'],
            'ability_name': data['abilities'][1]['ability']['name'],
            'base_experience': data['base_experience'],
            'attack_base_stat': data['stats'][1]['base_stat'],
            'hp_base_stat': data['stats'][2]['base_stat'],
            'defense_base_stat': data['stats'][2]['base_stat'],            
            'sprite': data['sprites']['front_shiny']
        }

        lucario_data.append(luc_dict)
        return lucario_data
get_luc_info(lucario_data)
print(get_luc_info(lucario_data))

# # #NEXT POKEMON: PSYDUCK
import requests
url = 'https://pokeapi.co/api/v2/pokemon/psyduck'
response = requests.get(url)
response.status_code

poke_data = response.json()
psyduck_data = poke_data

def get_psy_info(data):
    psyduck_data = []
    for psy in data:
        psy_dict = {
            'pokemon_name': data['forms'][0]['name'],
            'ability_name': data['abilities'][1]['ability']['name'],
            'base_experience': data['base_experience'],
            'attack_base_stat': data['stats'][1]['base_stat'],
            'hp_base_stat': data['stats'][2]['base_stat'],
            'defense_base_stat': data['stats'][2]['base_stat'],            
            'sprite': data['sprites']['front_shiny']
        }

        psyduck_data.append(psy_dict)
        return psyduck_data
get_psy_info(psyduck_data)
print(get_psy_info(psyduck_data))

# # #LAST POKEMON: DIALGA
import requests
url = 'https://pokeapi.co/api/v2/pokemon/dialga'
response = requests.get(url)
response.status_code

poke_data = response.json()
dialga_data = poke_data

def get_dia_info(data):
    dialga_data = []
    for dia in data:
        dia_dict = {
            'pokemon_name': data['forms'][0]['name'],
            'ability_name': data['abilities'][1]['ability']['name'],
            'base_experience': data['base_experience'],
            'attack_base_stat': data['stats'][1]['base_stat'],
            'hp_base_stat': data['stats'][2]['base_stat'],
            'defense_base_stat': data['stats'][2]['base_stat'],            
            'sprite': data['sprites']['front_shiny']
        }

        dialga_data.append(dia_dict)
        return dialga_data
get_dia_info(dialga_data)
print(get_dia_info(dialga_data))

# # #EXTRA POKEMON: UXIE
import requests
url = 'https://pokeapi.co/api/v2/pokemon/uxie'
response = requests.get(url)
response.status_code

poke_data = response.json()
uxie_data = poke_data

def get_ux_info(data):
    uxie_data = []
    for ux in data:
        ux_dict = {
            'pokemon_name': data['forms'][0]['name'],
            'ability_name': data['abilities'][0]['ability']['name'],
            'base_experience': data['base_experience'],
            'attack_base_stat': data['stats'][1]['base_stat'],
            'hp_base_stat': data['stats'][2]['base_stat'],
            'defense_base_stat': data['stats'][2]['base_stat'],            
            'sprite': data['sprites']['front_shiny']
        }

        uxie_data.append(ux_dict)
        return uxie_data
get_ux_info(uxie_data)
print(get_ux_info(uxie_data))
```
